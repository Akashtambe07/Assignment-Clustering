# K- Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
airline.head()
```

```python
airline3=airline.drop(['ID#'],axis=1)
```

```python
airline3
```

```python
airline3.info()
```

```python
airline3.isnull().sum()
```

```python
#Standardize the data
airline3_norm=StandardScaler().fit_transform(airline3)
```

```python
airline3_norm
```

```python
# by using WCSS criteria find k cluster
wcss=[]
for i in range(1,12):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(airline3_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,12),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
cluster=KMeans(6,random_state=42)
cluster.fit(airline3_norm)
```

```python
cluster.labels_
```

```python
airline3['cluster']=cluster.labels_
```

```python
airline3.groupby('cluster').agg(['mean']).reset_index()
```
